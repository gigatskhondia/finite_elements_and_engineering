```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

### ex. 9.1

```python
d=300
g=9.81
```

```python
omega=[0.1*(i+1) for i in range(int(2/0.1))]
```

```python
omega
```

```python
def wave_number(k, d, g, o):
    return g*k*tanh(k*d)-o**2
```

```python
x = Symbol('x')
```

```python
ans = [nsolve(wave_number(x, d, g, o), x, 1) for o in omega]
ans0 = [o**2/g for o in omega]
```

```python
for i in range(len(ans)):
    print(omega[i],abs(ans[i]-ans0[i])/ans0[i]*100<5)
```

```python
plt.plot(omega, ans, label='g*k*tanh(k*d)-omega**2')
plt.plot(omega, ans0, label='omega**2/g')
plt.xlabel('omega')
plt.ylabel('k')
plt.title('Wave number')
plt.legend(loc='best')
plt.show()
```
